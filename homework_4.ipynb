```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(4) 
y1 = np.array([1, 2, 3, 4])
y2 = np.array([5, 6, 7, 8])
y3 = np.array([9, 10, 11, 12])

plt.plot(x, y1, label='y1', marker='o')
plt.plot(x, y2, label='y2', marker='s')
plt.plot(x, y3, label='y3', marker='d')

plt.xticks(x, ('0', '1', '2', '3')) 
plt.show()

```

```python
import matplotlib.pyplot as plt
marks = {'Andy': 88, 'Amy':66,'James':90, 'Jules':55,'Arthur': 77}

students = list(marks.keys())
grades = list(marks.values())
import numpy as np 
x=np.array(["Andy","Amy", "James", "Jules", "Arthur"])
y=np.array([88,66,90,55,77])
for i in range(len(students)):
    plt.xlabel('Student Name')
    plt.ylabel('Grade')
    plt.bar(x,y)

plt.figure(figsize=(5,5))
plt.pie(grades, labels=students, autopct='%1.1f%%')
plt.title('All Students Grades')
plt.legend(title='Student Names')

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,1,2,3])
y=np.array([3,5,7,9])

plt.subplot(2,3,1)
plt.plot(x,y)
plt.grid()
plt.title('Grid')

x=np.array([0,2,4,6])
y=np.array([2,4,6,8])

plt.subplot(2,3,2)
plt.bar(x,y)
plt.title('Bar Graph')

x=np.random.normal(170,10,250)

plt.subplot(2,3,3)
plt.hist(x)
plt.title('Histogram')

x=np.array([2,5,8,11,14])
y=np.array([10,7,9,4,17])

plt.subplot(2,3,4)
plt.scatter(x,y)
plt.title('Scatterplot')

x=np.array([1,2,5,9])

plt.subplot(2,3,5)
plt.pie(x)
plt.title('Piechart')

x1=np.array([1,3,5,9,10,13])
y1=np.array([5,6,7,8,14,16])
x2=np.array([0,2,4,6,8,10])
y2=np.array([2,4,6,8,12,14])

plt.subplot(2,3,6)
plt.plot(x1,y1,x2,y2)
plt.title('Multiple Lines')

plt.subplots_adjust(hspace=0.4, wspace=0.4)
```
